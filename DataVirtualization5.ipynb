<h1>Data Analysis and Virtualization with Heat Map and Line Graph</h1>

```python
import seaborn as sns
fligths = sns.load_dataset('flights')
df = fligths.copy()
df.head()
```

```python
df['passengers'].describe()
```

```python
df = df.pivot(index='month', columns='year', values='passengers')
df.head()
```

```python
df
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df, annot=True, fmt='d')
```

```python
sns.heatmap(df, annot=True, fmt='d', linewidths=.2)
```

Line Graph

```python
fmri = sns.load_dataset('fmri')
df1 = fmri.copy()
df1.head()
```

```python
df1.describe().T
```

```python
df1.groupby("timepoint")["signal"].count()
```

```python
df1.groupby("signal").count()
```

```python
sns.lineplot(x="timepoint", y="signal", data=df1)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=df1)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="event", data=df1)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="event", markers=True, dashes=False, data=df1)
```
