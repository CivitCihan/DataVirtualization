<h1>Data Analysis and Virtualization with Violin and Correlation</h1>

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.catplot(y="total_bill", hue="time", kind="violin", data=df)
```

```python
sns.catplot(x="day", y="total_bill", hue="time", kind="violin", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
```

```python
sns.scatterplot(x="tip", y="total_bill", hue="smoker", style="smoker", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="day", data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", size="size", hue="size", data=df)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill", y="tip", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=df)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="smoker")
```

```python
sns.pairplot(df, hue="smoker", markers=["D", "s"])
```

```python
sns.pairplot(df, hue="time", kind="reg")
```
