<h1>Data Analysis and Virtualization with Bar Plot</h1>

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df[["cut", "carat"]].value_counts()
```

```python
from pandas.api.types import CategoricalDtype
```

```python
df.cut.astype(CategoricalDtype(ordered=True))
```

```python
cut_cate = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut.astype(CategoricalDtype(categories= cut_cate, ordered=True))
```

```python
df["cut"].value_counts().plot.barh().set_title("Cut values class frequencies")
```

```python
sns.barplot(x="cut", y=df.cut.index, data=df)
```

```python
sns.barplot(x="price", y="cut", hue="color", data=df)
```

```python
df.groupby(["cut", "color"])["price"].mean()
```

```python
sns.catplot(x="cut", y="price", hue="color", data=df)
```

```python
# Histogram, distplot or histplot
sns.histplot(df.price, bins=500, kde= False)
```

```python
sns.distplot(df.price, hist= False)
```

```python
(sns
 .FacetGrid(df,
              hue = "cut",
              height = 5,
              xlim = (0, 10000))
.map(sns.kdeplot, "price", fill = True)
.add_legend()
);
```

```python
sns.catplot(x="cut", y="price", hue="color", data=df, kind="point")
```
